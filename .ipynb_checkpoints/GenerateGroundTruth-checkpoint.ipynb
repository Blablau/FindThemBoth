```python
import csv
from shutil import copyfile

with open('./FolderFrames/GT_AllShows.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            
            #print(f'\t{row[0]} works in the {row[1]} department, and was born in {row[2]}.')
            framenumber = int(row[0])
            hasKermit = int(row[1])
            hasMrsPiggy = int(row[2])
            shownumber = 1
            if framenumber == 1:
                continue
            if framenumber <= 1500:
                shownumber = 1
            elif framenumber <= 3000:
                shownumber = 2
            else:
                shownumber = 3
            try:
                if hasKermit == 1:
                    if framenumber % 1500 == 0:
                        copyfile('./FolderFrames/AllShows/show' + str(shownumber) + '_' + str(1500) + '.jpg', './FolderFrames/YesKermit/show' + str(shownumber) + '_' + str(1500) + '.jpg')
                    else:
                        copyfile('./FolderFrames/AllShows/show' + str(shownumber) + '_' + str(framenumber%1500) + '.jpg', './FolderFrames/YesKermit/show' + str(shownumber) + '_' + str(framenumber%1500) + '.jpg')
                else:
                    if framenumber % 1500 == 0:
                        copyfile('./FolderFrames/AllShows/show' + str(shownumber) + '_' + str(1500) + '.jpg', './FolderFrames/NoKermit/show' + str(shownumber) + '_' + str(1500) + '.jpg')
                    else:
                        copyfile('./FolderFrames/AllShows/show' + str(shownumber) + '_' + str(framenumber%1500) + '.jpg', './FolderFrames/NoKermit/show' + str(shownumber) + '_' + str(framenumber%1500) + '.jpg')
                if hasMrsPiggy == 1:
                    if framenumber % 1500 == 0:
                        copyfile('./FolderFrames/AllShows/show' + str(shownumber) + '_' + str(1500) + '.jpg', './FolderFrames/YesMrsPiggy/show' + str(shownumber) + '_' + str(1500) + '.jpg')
                    else:
                        copyfile('./FolderFrames/AllShows/show' + str(shownumber) + '_' + str(framenumber%1500) + '.jpg', './FolderFrames/YesMrsPiggy/show' + str(shownumber) + '_' + str(framenumber%1500) + '.jpg')
                else:
                    if framenumber % 1500 == 0:
                        copyfile('./FolderFrames/AllShows/show' + str(shownumber) + '_' + str(1500) + '.jpg', './FolderFrames/NoMrsPiggy/show' + str(shownumber) + '_' + str(1500) + '.jpg')
                    else:
                        copyfile('./FolderFrames/AllShows/show' + str(shownumber) + '_' + str(framenumber%1500) + '.jpg', './FolderFrames/NoMrsPiggy/show' + str(shownumber) + '_' + str(framenumber%1500) + '.jpg')
            except:
                print("line_count" + str(line_count))
            line_count += 1
    print(f'Processed {line_count} lines.')
```
